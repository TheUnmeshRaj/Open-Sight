```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset_cleaned.csv")

KEEP_COLS = [
    "FIR_DATE",
    "Latitude",
    "Longitude",
    "CrimeGroup_Name",
    "VICTIM COUNT"
]

df = df[KEEP_COLS].copy()
```

```python
LAT_MIN = 12.70
LAT_MAX = 13.30
LON_MIN = 77.30
LON_MAX = 78.00
LAT_GRIDS = 50
LON_GRIDS = 50
SEQ_LEN = 12
START_DATE = "'2019-01-01'"
START_SELECT_DATE = "'2019-02-01'"
END_DATE = "'2025-12-31'"
TRAIN_VAL_DATE = "'2024-01-01'"
VAL_TEST_DATE = "'2025-01-01'"
```

```python
df = df.dropna(subset=["Latitude", "Longitude", "FIR_DATE"])

df["FIR_DATE"] = pd.to_datetime(df["FIR_DATE"])

# Keep only Bangalore bounding box
df = df[
    (df["Latitude"] >= LAT_MIN) & (df["Latitude"] <= LAT_MAX) &
    (df["Longitude"] >= LON_MIN) & (df["Longitude"] <= LON_MAX)
]
print(f"Remaining rows after cleaning: {len(df)}")
```

```python
lat_bins = np.linspace(LAT_MIN, LAT_MAX, LAT_GRIDS + 1)
lon_bins = np.linspace(LON_MIN, LON_MAX, LON_GRIDS + 1)

df["grid_x"] = np.digitize(df["Latitude"], lat_bins) - 1
df["grid_y"] = np.digitize(df["Longitude"], lon_bins) - 1
```

```python
df["day"] = df["FIR_DATE"].dt.day
df["month"] = df["FIR_DATE"].dt.month
df["year"] = df["FIR_DATE"].dt.year
df["day_of_week"] = df["FIR_DATE"].dt.weekday
df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)
```

```python
agg = (
    df.groupby(["FIR_DATE", "grid_x", "grid_y"])
      .agg(
          crime_count=("CrimeGroup_Name", "count"),
          victim_count=("VICTIM COUNT", "sum")
      )
      .reset_index()
)

agg
```

```python
all_dates = pd.date_range(
    agg["FIR_DATE"].min(),
    agg["FIR_DATE"].max(),
    freq="D"
)

grid_index = pd.MultiIndex.from_product(
    [all_dates, range(LAT_GRIDS), range(LON_GRIDS)],
    names=["FIR_DATE", "grid_x", "grid_y"]
)

full = (
    agg.set_index(["FIR_DATE", "grid_x", "grid_y"])
       .reindex(grid_index, fill_value=0)
       .reset_index()
)
```

```python
full = full.sort_values(["grid_x", "grid_y", "FIR_DATE"])

for lag in [1, 3, 7, 14, 30]:
    full[f"crime_lag_{lag}"] = (
        full.groupby(["grid_x", "grid_y"])["crime_count"]
            .shift(lag)
    )

full = full.dropna()
```

```python
full["crime_occurred"] = (full["crime_count"] > 0).astype(int)

full
```

```python
type_df = (
    df.groupby(["FIR_DATE", "grid_x", "grid_y", "CrimeGroup_Name"])
      .size()
      .reset_index(name="count")
)

type_df = type_df[type_df["count"] > 0]
type_df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

type_train["crime_label"] = le.fit_transform(type_train["CrimeGroup_Name"])
type_val["crime_label"] = le.transform(type_val["CrimeGroup_Name"])
```

```python
valid_classes = set(type_train["CrimeGroup_Name"].unique())
type_val["CrimeGroup_Name"] = type_val["CrimeGroup_Name"].apply(
    lambda x: x if x in valid_classes else "OTHER"
)
if "OTHER" not in valid_classes:
    type_train = pd.concat([
        type_train,
        pd.DataFrame({
            "CrimeGroup_Name": ["OTHER"],
            **{c: [0] for c in FEATURES}
        })
    ], ignore_index=True)

le = LabelEncoder()
type_train["crime_label"] = le.fit_transform(type_train["CrimeGroup_Name"])
type_val["crime_label"] = le.transform(type_val["CrimeGroup_Name"])
```

```python
from xgboost import XGBClassifier

model2 = XGBClassifier(
    objective="multi:softprob",
    num_class=len(le.classes_),
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model2.fit(
    type_train[FEATURES],
    type_train["crime_label"]
)
```

```python
future_dates = pd.date_range("2026-01-01", "2026-01-31", freq="D")

future = pd.MultiIndex.from_product(
    [future_dates, range(LAT_GRIDS), range(LON_GRIDS)],
    names=["FIR_DATE", "grid_x", "grid_y"]
).to_frame(index=False)

future["day_of_week"] = future["FIR_DATE"].dt.weekday
future["is_weekend"] = (future["day_of_week"] >= 5).astype(int)
future["month"] = future["FIR_DATE"].dt.month
```

```python
last_known = full[full["FIR_DATE"] <= "2025-12-31"]
lag_features = {}

for lag in [1, 7, 30]:
    lag_map = (
        last_known
        .sort_values("FIR_DATE")
        .groupby(["grid_x", "grid_y"])
        .tail(lag)
        .groupby(["grid_x", "grid_y"])["crime_count"]
        .mean()
        .reset_index()
        .rename(columns={"crime_count": f"crime_lag_{lag}"})
    )
    lag_features[lag] = lag_map
```

```python
for lag, lag_df in lag_features.items():
    future = future.merge(
        lag_df,
        on=["grid_x", "grid_y"],
        how="left"
    )
```

```python
future[["crime_lag_1", "crime_lag_7", "crime_lag_30"]] = (
    future[["crime_lag_1", "crime_lag_7", "crime_lag_30"]].fillna(0)
)
```

```python
print(future[FEATURES].head())
```

```python
future["crime_probability"] = model1.predict_proba(
    future[FEATURES]
)[:, 1]

future["is_unsafe"] = future["crime_probability"] > THRESHOLD
```

```python
unsafe = future[future["is_unsafe"]].copy()

crime_probs = model2.predict_proba(unsafe[FEATURES])
unsafe["predicted_crime"] = le.inverse_transform(
    crime_probs.argmax(axis=1)
)

future = future.merge(
    unsafe[["FIR_DATE","grid_x","grid_y","predicted_crime"]],
    on=["FIR_DATE","grid_x","grid_y"],
    how="left"
)
```

```python
lat_size = (LAT_MAX - LAT_MIN) / LAT_GRIDS
lon_size = (LON_MAX - LON_MIN) / LON_GRIDS

future["lat"] = LAT_MIN + (future["grid_x"] + 0.5) * lat_size
future["lon"] = LON_MIN + (future["grid_y"] + 0.5) * lon_size
```
